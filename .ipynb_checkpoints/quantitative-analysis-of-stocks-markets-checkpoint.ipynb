### Quantitative Analysis of Stock Market

```python
# https://thecleverprogrammer.com/2024/01/15/quantitative-analysis-of-stock-market-using-python/
#data link https://statso.io/quantitative-analysis-case-study/
import pandas as pd

stocks_data = pd.read_csv('stocks.csv')
```

```python
stocks_data.head()
```

### Descriptive statistics
Descriptive Statistics will provide summary statistics for each stock in the dataset. Weâ€™ll look at measures such as mean, median, standard deviation, and more for the Close prices:

```python
descriptive_stats = stocks_data.groupby('Ticker')['Close'].describe()

print(descriptive_stats)
```

### Time Series Analysis

```python
# Time Series Analysis
stocks_data['Date'] = pd.to_datetime(stocks_data['Date'])
pivot_data = stocks_data.pivot(index = 'Date', columns  = 'Ticker',
                              values = 'Close')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Set up the plot
plt.figure(figsize=(12, 6))

# Plot each ticker using Seaborn
for column in pivot_data.columns:
    sns.lineplot(data=pivot_data, x=pivot_data.index, y=column, label=column)

# Customize the plot
plt.title('Time Series of Closing Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend(title='Ticker')

# Show the plot
plt.tight_layout()
plt.show()
```

### Volatility

```python
volatility = pivot_data.std().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
ax = sns.barplot(x=volatility.index, y=volatility.values)

plt.title('Volatility of Closing Prices (Standard Deviation)')
plt.xlabel('Ticker')
plt.ylabel('Standard Deviation')
plt.xticks(rotation=45)

# Adding values on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}',  
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 5),
                textcoords='offset points')

plt.tight_layout()
plt.show()
```

The bar chart and the accompanying data show the volatility (measured as standard deviation) of the closing prices for each stock. Hereâ€™s how they rank in terms of volatility:

**NFLX**: Highest volatility with a standard deviation of approximately 18.55.

**MSFT**: Next highest, with a standard deviation of around 17.68.

**AAPL**: Lower volatility compared to NFLX and MSFT, with a standard deviation of about 7.36.

**GOOG**: The least volatile in this set, with a standard deviation of approximately 6.28.
It indicates that NFLX and MSFT stocks were more prone to price fluctuations during this period compared to AAPL and GOOG.

### Correlation Analysis
Next, weâ€™ll perform a Correlation Analysis to understand how the stock prices of these companies are related to each other:

```python
# Correlation Analysis
correlation_matrix = pivot_data.corr()

# Create figure
plt.figure(figsize=(10, 8))

# Create heatmap with Seaborn
ax = sns.heatmap(correlation_matrix,
                 annot=True,  # Show correlation values in cells
                 fmt=".2f",  # Format to 2 decimal places
                 cmap='Blues',  # Blue color scale
                 vmin=-1, vmax=1,  # Set color scale bounds
                 square=True)  # Make cells square

# Customize plot
plt.title('Correlation Matrix of Closing Prices')
plt.xlabel('Ticker')
plt.ylabel('Ticker')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show plot
plt.show()
```

The heatmap above displays the correlation matrix of the closing prices of the four stocks (AAPL, GOOG, MSFT, NFLX). Hereâ€™s what the correlation coefficients suggest:

- Values close to +1 indicate a strong positive correlation, meaning that as one stockâ€™s price increases, the other tends to increase as well.

- Values close to -1 indicate a strong negative correlation, where one stockâ€™s price increase corresponds to a decrease in the other.

- Values around 0 indicate a lack of correlation.

- 
From the heatmap, we can observe that there are varying degrees of positive correlations between the stock prices, with some pairs showing stronger correlations than others. For instance, AAPL and MSFT seem to have a relatively higher positive correlation.


### Comparative analysis 
 In this step, weâ€™ll compare the performance of different stocks based on their returns over the period. Weâ€™ll calculate the percentage change in closing prices from the start to the end of the period for each stock:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate percentage change
percentage_change = ((pivot_data.iloc[-1] - pivot_data.iloc[0]) / pivot_data.iloc[0]) * 100

# Create DataFrame for Seaborn
change_df = percentage_change.reset_index()
change_df.columns = ['Ticker', 'Percentage Change']

# Create plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=change_df, x='Ticker', y='Percentage Change', palette='viridis')

# Add values on top of bars
for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}%", 
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', 
                xytext=(0, 9), 
                textcoords='offset points')

# Customize plot
plt.title('Percentage Change in Closing Prices', pad=20, fontsize=14)
plt.ylabel('Percentage Change (%)', labelpad=10)
plt.xlabel('Ticker', labelpad=10)
plt.xticks(rotation=45)

# Add zero line for reference
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')

# Adjust layout
plt.tight_layout()
plt.show()
```

The bar chart and the accompanying data show the percentage change in the closing prices of the stocks from the start to the end of the observed period:

**MSFT**: The highest positive change of approximately 16.10%.

**AAPL**: Exhibited a positive change of approximately 12.23%. It indicates a solid performance, though slightly lower than MSFTâ€™s.

**GOOG**: Showed a slight negative change of about -1.69%. It indicates a minor decline in its stock price over the observed period.

**NFLX**: Experienced the most significant negative change, at approximately -11.07%. It suggests a notable decrease in its stock price during the period.

### Daily Risk Vs. Return Analysis
To perform a Risk vs. Return Analysis, we will calculate the average daily return and the standard deviation of daily returns for each stock. The standard deviation will serve as a proxy for risk, while the average daily return represents the expected return.

We will then plot these values to visually assess the risk-return profile of each stock. Stocks with higher average returns and lower risk (standard deviation) are generally more desirable, but investment decisions often depend on the investorâ€™s risk tolerance:

```python
# Calculate returns and risk metrics
daily_returns = pivot_data.pct_change().dropna()
risk_return_df = pd.DataFrame({
    'Risk': daily_returns.std(),
    'Average Daily Return': daily_returns.mean()
})

# Format values for display
risk_return_df['Risk_pct'] = risk_return_df['Risk'].apply(lambda x: f"{x:.2%}")
risk_return_df['Return_pct'] = risk_return_df['Average Daily Return'].apply(lambda x: f"{x:.2%}")

# Set style with adjusted parameters
sns.set_style("whitegrid", {'grid.linestyle': '--', 'grid.alpha': 0.4})
plt.figure(figsize=(10, 7))  # Increased height

# Create scatter plot with adjusted margins
ax = sns.scatterplot(
    data=risk_return_df,
    x='Risk',
    y='Average Daily Return',
    s=120,
    color='#2b7bba',  # Professional blue
    alpha=0.9,
    edgecolor='w',  # White border for markers
    linewidth=0.5
)

# Title with manual positioning to avoid overlap
plt.title('Risk vs. Return Analysis', 
          y=1.05,  # Pushes title upward
          fontsize=14,
          pad=20)

# Smart annotation placement
for label, x, y, risk_val, return_val in zip(
    risk_return_df.index,
    risk_return_df['Risk'],
    risk_return_df['Average Daily Return'],
    risk_return_df['Risk_pct'],
    risk_return_df['Return_pct']
):
    # Dynamic offset based on position
    x_offset = 15 if x < risk_return_df['Risk'].median() else -15
    y_offset = 15 if y < risk_return_df['Average Daily Return'].median() else -15
    ha = 'left' if x_offset > 0 else 'right'
    va = 'bottom' if y_offset > 0 else 'top'
    
    ax.annotate(
        f"{label}\nÏƒ: {risk_val}\nÎ¼: {return_val}",
        xy=(x, y),
        xytext=(x_offset, y_offset),
        textcoords='offset points',
        fontsize=9,
        ha=ha,
        va=va,
        bbox=dict(
            boxstyle='round,pad=0.4',
            fc='white',
            alpha=0.85,
            ec='lightgray',
            lw=0.5
        ),
        linespacing=1.5
    )

# Format axes as percentages
ax.xaxis.set_major_formatter(plt.FuncFormatter('{0:.0%}'.format))
ax.yaxis.set_major_formatter(plt.FuncFormatter('{0:.0%}'.format))

# Axis labels with padding
plt.xlabel('Risk (Ïƒ)', labelpad=12)
plt.ylabel('Avg Daily Return (Î¼)', labelpad=12)

# Adjust plot area to make space
plt.subplots_adjust(top=0.85)  # Extra space at top

# Custom grid
ax.yaxis.grid(True, linestyle='--', alpha=0.4)
ax.xaxis.grid(False)
sns.despine()

# Equal aspect ratio if scales are comparable
x_range = risk_return_df['Risk'].max() - risk_return_df['Risk'].min()
y_range = risk_return_df['Average Daily Return'].max() - risk_return_df['Average Daily Return'].min()
if 0.5 < x_range/y_range < 2:
    ax.set_aspect('equal')

plt.tight_layout()
plt.show()
```
